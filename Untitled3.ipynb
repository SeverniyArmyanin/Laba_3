# Задание 1. Построение простого графика. 

Напишите код, который строит график функции $y = x^2$ на интервале [−10, 10].
Добавьте заголовок, подписи осей и сетку.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
y = x**2

plt.plot(x, y)
plt.title("График функции y = x²")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

# Задание 2. Настройка стилей и цветов. 

Постройте три линии на одном графике:

$y = x$ (синяя, пунктирная линия),

$y = x^2$ (зеленая, штрих-пунктирная линия),

$y = x^3$ (красная, сплошная линия).

Добавьте легенду и сделайте оси одинакового масштаба.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 500)

plt.plot(x, x, 'b--', label='y = x')        # синяя пунктирная линия
plt.plot(x, x**2, 'g-.', label='y = x²')    # зелёная штрих-пунктирная линия
plt.plot(x, x**3, 'r-', label='y = x³')     # красная сплошная линия

plt.ylim(-10, 10)
plt.title("Графики: y = x, y = x², y = x³")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.axhline(0, color='black', linewidth=1.2)  # ось X
plt.axvline(0, color='black', linewidth=1.2)  # ось Y

plt.show()
```

# Задание 4. Гистограмма распределения. 

Сгенерируйте 1000 случайных чисел из нормального распределения с
параметрами $μ = 0$, $σ = 1$ и постройте их гистограмму с 30 бинами. Добавьте
вертикальную линию в среднем значении.

```python
import numpy as np
import matplotlib.pyplot as plt

# генерация данных из нормального распределения
data = np.random.normal(loc=0, scale=1, size=1000)

# вычисление среднего
mean_value = np.mean(data)

# построение гистограммы
plt.hist(data, bins=30, color='#00FF7F', edgecolor='black')

# вертикальная линия в среднем значении
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f'Среднее = {mean_value:.2f}')

plt.title("Гистограмма нормального распределения")
plt.xlabel("Значения")
plt.ylabel("Частота")
plt.legend()
plt.grid(True)

plt.show()
```

# Задание 7. Трехмерный график поверхности.

Используя **mpl_toolkits.mplot3d**, постройте 3D-график функции $z = sin(√x^2 + y^2)$
на сетке значений $x$, $y$ в диапазоне [−5, 5].

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # подключаем 3D-поддержку

# Создание сетки значений
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Вычисление значений функции
r = np.sqrt(x**2 + y**2)
z = np.sin(r)

# Создание 3D-графика
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Добавление заголовков
ax.set_title('z = sin(√(x² + y²))')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

# Задание 8. Множественные подграфики (subplots).

Постройте четыре графика в одной фигуре (2 × 2 сетка):

1. Линейный график $y = x$
2. Парабола $y = x^2$
3. Синус $y = sin(x)$
4. Косинус $y = cos(x)$

Добавьте заголовки к каждому подграфику.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y_linear = x
y_parabola = x**2
y_sin = np.sin(x)
y_cos = np.cos(x)

# создание фигуры и 2×2 сетки подграфиков
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# первый 
axs[0, 0].plot(x, y_linear)
axs[0, 0].set_title('y = x')
axs[0, 0].grid(True)

# второй
axs[0, 1].plot(x, y_parabola)
axs[0, 1].set_title('y = x²')
axs[0, 1].grid(True)

# третий
axs[1, 0].plot(x, y_sin)
axs[1, 0].set_title('y = sin(x)')
axs[1, 0].grid(True)

# четвёртый
axs[1, 1].plot(x, y_cos)
axs[1, 1].set_title('y = cos(x)')
axs[1, 1].grid(True)

# автоматическая подстройка расстояний между графиками
plt.tight_layout()

plt.show()
```

# Индивидуальное задание №3. Вариант 8. Площадь под многочленом 

Определите площадь под кривой:

$$f(x) = cos(x)$$

на отрезке $[0, π/2]$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

# вычисляем данные
f = lambda x: 3*x**4 - 2*x**3 + x**2 - 5
x = np.linspace(-1, 1, 500)
y = f(x)
integral = integrate.quad(f, -1, 1)[0]

# график
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'r', lw=2, label='f(x) = 3x⁴ - 2x³ + x² - 5')
plt.fill_between(x, y, y.min(), color='gray', alpha=0.3)

plt.title('График функции и площадь под кривой')
plt.xlabel('x'); plt.ylabel('f(x)')
plt.axhline(0, c='black', lw=0.5)
plt.axvline(0, c='black', lw=0.5)
plt.grid(ls='--', alpha=0.7)
plt.legend()
plt.text(0, -4, f'∫f(x)dx = {integral:.2f}', ha='center', va='center', 
         bbox=dict(facecolor='white', alpha=0.8))
plt.show()
```

# Индивидуально задание №4. Вариант 8. Волнообразная поверхность 

Визуализируйте функцию:

$$f(x, y) = sin(x^2 + y^2)$$

на $x, y ∈ [−2π, 2π].$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# сетка в диапазоне [-2π, 2π]
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)
y = np.linspace(-2 * np.pi, 2 * np.pi, 400)
X, Y = np.meshgrid(x, y)

Z = np.sin(X**2 + Y**2)

# создаем 3D фигуру
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# строим 3D график
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# 5. Добавляем заголовок и под
ax.set_title('График функции f(x, y) = sin(x² + y²)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')

# 6. Добавляем цветовую карту
fig.colorbar(surf)

# 7. Показать график
plt.show()
```
